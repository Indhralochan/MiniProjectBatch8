```python
df['artist_genres'] = df['artist_genres'].astype(str)
```

```python
df['artist_genres']=df['artist_genres'].apply(lambda x: ''.join(x.split(';')))
```

```python
df.head()
```

```python
df['tag']=df['track_name'].astype(str)+' '+df['album_name'].astype(str)+' '+df['artists_names'].astype(str)+' '+df['artist_genres'].astype(str)
```

```python
df.head()
```

```python
new_df=df[['track_id','track_name','popularity','album_release_date','tag']]
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tag']).toarray()
```

```python
vector.shape
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
def recommend(song):
    index = new_df[new_df['track_name'] == song].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:11]:
        print(new_df.iloc[i[0]].track_name)
```

```python
recommend("Shot in the Dark")
```

```python
recommend("The Ultimate Sin")
```

```python
def recommend_popular_similar_songs(new_df, similarity, num_recommendations=10):
    mean_popularity = new_df['popularity'].mean()
    recommendation_scores = pd.DataFrame({
        'track_name': new_df['track_name'],
        'popularity': new_df['popularity'],
        'similarity_score': similarity.mean(axis=1)
    })
    recommendation_scores['combined_score'] = (recommendation_scores['popularity'] + recommendation_scores['similarity_score']) / 2
    recommended_songs = recommendation_scores.sort_values(by='combined_score', ascending=False).head(num_recommendations)
    return recommended_songs
```

```python
recommended_songs = recommend_popular_similar_songs(new_df, similarity, num_recommendations=100)
pickle_file_path = 'recommended_songs.pkl'
with open(pickle_file_path, 'wb') as pickle_file:
    pk.dump(recommended_songs, pickle_file)
print(f'Recommended songs saved to {pickle_file_path}')
```
